Importação das bibliotecas

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

Carregamento dos dados em um Dataframe

````python
df = pd.read_csv("Profissao- Cientista de Dados_M10_support material.csv")
df
````

Análise da variável posse_de_imovel em relação à renda

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda",
                   data=df, dodge=True, errorbar=('ci', 95))
````

Análise da variável posse_de_imovel por sexo em relação à renda

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda", hue='sexo',
                   data=df, dodge=True, errorbar=('ci', 95))

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

Análise da variável posse_de_veiculo em relação à renda

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=df, dodge=True, errorbar=('ci', 95))
````

Análise da variável posse_de_veiculo por sexo em relação à renda

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda", hue='sexo',
                   data=df, dodge=True, errorbar=('ci', 95))

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

Analisando apenas as variáveis qualitativas posse_de_imovel e posse_de_veiculo, a indicação do pointplot é que são ambas muito significativas para predição de renda, porém, aplicado agrupamento por sexo, a variável posse_de_veiculo apresenta um comportamento diferente. Para a observação relacionada ao sexo para predição da renda parecem ser necessárias outras considerações, afim de evitar viesamento da análise
