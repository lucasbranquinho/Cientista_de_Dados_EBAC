```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
df = pd.read_csv("SINASC_RO_2019.csv")
df.describe()
```

```python
sns.boxplot(data= df,
            y='IDADEMAE')
```

####  2) Faça um boxplot do peso do bebê

```python
df[['PESO']].boxplot()
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
sns.displot(data=df,
            x='IDADEMAE',
            alpha= .25,
            bins= 20,
            kde= True,
            element= 'step'
            )
```

#### 4) Faça um histograma do peso do bebê

```python
sns.displot(data=df,
            x='PESO',
            alpha= .25,
            bins= 20,
            kde= True,
            element= 'step')
```
